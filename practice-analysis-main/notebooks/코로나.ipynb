```python
# type확인

covid_df =covid19.groupby("확진일")["연번"].count().reset_index()
covid_df
```

```python
covid_df.columns = ["ds", "y"]
covid_df
```

```python
from  prophet import Prophet
```

```python
m = Prophet()
m.fit(covid_df)
```

```python
future = m.make_future_dataframe(periods=30)
future.tail()
```

```python
# 예측 하기
# 추측값 넣기
forecast = m.predict(future)
forecast.tail()
```

```python
m.plot(forecast)
```

```python
from prophet.plot import add_changepoints_to_plot
fig = m.plot(forecast)
a = add_changepoints_to_plot(fig.gca(),m,forecast)
```
